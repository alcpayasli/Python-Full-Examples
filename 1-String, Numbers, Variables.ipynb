## Basic Strings

```python
my_string = "Hello World!"
my_string2 = 'Hello World.'
print(my_string)
print(my_string2)
```

```python
print(len(my_string))
print(len(my_string2))
```

```python
a = "" # length 0
b = " " # lenght 1
print(len(a))
print(len(b))
print(type(""))
print(type(" "))
```

```python
# Concat Strings
print("www." + "kaggle" + ".com")
print("www." "kaggle" ".com")
```

```python
print("w"*3 + "." + "kaggle" ".com")
```

```python
print("kaggle"*3)
```

```python
print("Machine" + " "*5 + "Learning")
```
