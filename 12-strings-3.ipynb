# strip, lstrip, rstrip

```python
text = """Python is a high-level, general-purpose programming language. Its design philosophy emphasizes code readability with the use of significant indentation.[31]

Python is dynamically typed and garbage-collected. It supports multiple programming paradigms, including structured (particularly procedural), object-oriented and functional programming. It is often described as a "batteries included" language due to its comprehensive standard library.
"""
```

```python
for i in text.split():
    i = i.strip(".")
    i = i.strip()
    i = i.strip(",")
    i = i.strip("(")
    i = i.strip(")")
    i = i.strip('"')
    print(i)
```

```python
obj = ",Hello world"
obj.lstrip(",")
```

```python
a = "Python Programming Langugage"
a.rstrip("gage")
```
