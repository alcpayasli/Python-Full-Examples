```python
# Example Bool Operators and Condition
username = input("Input username:")
password = input("Input Password:")
if (not username) or (not password):
    print("Username or password can not be empty")
elif (username == "Python" and password == "Programming"):
    print("Welcome to my world!")
else:
    print("Username or password incorrect!!!")
```

```python
# Exam Grade Sample
grade = int(input("Input grade of student:"))
if(grade>100 or grade<0):
    print('Please enter correct values')
elif(grade>=90):
    print("AA")
elif(grade>=80):
    print("BA")
elif(grade>=70):
    print("BB")
elif(grade>=60):
    print("CB")
elif(grade>=50):
    print("CC")
else:
    print("DD")
```

```python
# Exam Grade Sample (2.way)
grade = int(input("Input grade of student:"))
if(grade>100 or grade<0):
    print('Please enter correct values')
elif(grade>=90 <=100):
    print("AA")
elif(grade>=80 <=89):
    print("BA")
elif(grade>=70 <=79):
    print("BB")
elif(grade>=60 <=69):
    print("CB")
elif(grade>=50 <=59):
    print("CC")
else:
    print("DD")
```

```python
# Exam Grade Sample (3.way)
grade = int(input("Input grade of student:"))
if(grade>100 or grade<0):
    print('Please enter correct values')
elif(90<=grade<=100):
    print("AA")
elif(80<=grade<=89):
    print("BA")
elif(70<=grade<=79):
    print("BB")
elif(60<=grade<=69):
    print("CB")
elif(50<=grade<=59):
    print("CC")
else:
    print("DD")
```
