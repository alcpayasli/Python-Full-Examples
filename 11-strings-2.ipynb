```python
name = input("Enter your name in lower case:")
if not name.islower():
    print("Please Enter your name in lower case")
else:
    print(name)
```

```python
mesage = input("Enter the message:")
for i in mesage.split():
    if i.isupper():
        print("Do not use words consisting entirely of capital letters")
        break
else:
    print(mesage)
```
