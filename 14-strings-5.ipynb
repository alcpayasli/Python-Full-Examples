# maketrans, translate

```python
sentence = "İstanbul Büyükşehir Belediyesi"
source = 'İüş'
target = 'Ius'
trans = str.maketrans(source, target)
print(sentence.translate(trans))
```

```python
print(trans)
```

```python
print(chr(304))
print(chr(73))
```

```python
for i,j in trans.items():
    print(chr(i), chr(j))
```

```python
text = 'Python is a high-level, general-purpose programming language. Its design philosophy emphasizes code readability with the use of significant indentation.'
source = 'aeiou'
trans = str.maketrans('', '', source)
print(text.translate(trans))
```

```python
print(trans)
```

```python
name = "Python Programming Language"
source = 'PL'
target = 'pl'
deleting = ' '
trans = str.maketrans(source, target, deleting)
print(name.translate(trans))
```
