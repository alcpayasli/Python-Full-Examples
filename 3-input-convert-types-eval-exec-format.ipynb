## Convert Type

```python
a = 345678
print(type(a))
```

```python
a = str(a) # integer convert to string
print(type(a))
print(len(a))
```

```python
b = "23238123"
print(type(b))
b = int(b)  # string convert to integer
print(type(b))
```

```python
print(type(a))
a = float(a) # string convert to float
print(type(a))
print(a)
```

```python
c = 324.56
print(type(c))
c = int(c)  # float convert to integer
print(c)
print(type(c))
```

```python
a = 12
a = complex(a) # integer convert to complex
print(a)
```

```python
b = 13.4 # float convert to complex
print(complex(b))
```
