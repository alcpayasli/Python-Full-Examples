# '%s'

```python
print("%s programming language" %"Python")
```

```python
language = "Python"
for i, j in enumerate(language):
    print("%s. char: %s" %(i,j))
```

```python
for i in range(10):
    print("%s" %i)
```

```python
for i in range(10):
    print("%%%s" %i)
```

```python
for i in dir(str):
    print("%15s" %i)
```

```python
for i in dir(str):
    print("|%15s|" %i)
```

```python
for i in dir(str):
    print("|%s|" %i.rjust(15))
```

```python
for i in dir(str):
    print("|%-15s|" %i)
```

```python
for i in dir(str):
    print("|%s|" %i.ljust(15))
```

# '%o-x-X'

```python
print("%s is octal value: %o" %(10,10))
```

```python
print("%s is hex value: %x" %(10,10))
```

```python
print("%s is hex value: %X" %(10,10))
```

# '%f-c'

```python
print("It's %f dolars."%(13.72))
```

```python
print("It's %.2f dolars."%(13.72))
```

```python
print('%c is only one character.'%'a')
```

```python
print('%c' %65)
```

```python
for i in range(128):
    print("%s...:%c" %(i,i))
```

```python
# Example
for i,j in enumerate(dir(str)):
    if i%4 == 0:
        print("\n", end = "")
    print("%-20s" % j, end = "")
```

```python
for i in range(20):
    print("%-5s%-5o%-5x" %(i,i,i))
```

```python
for i in range(20):
    print("%(value)-5s%(value)-5o%(value)-5x" %({"value": i}))
```
