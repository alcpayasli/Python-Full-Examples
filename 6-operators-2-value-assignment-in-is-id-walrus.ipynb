# in,is and id operators

```python
word = "Python Programming Language"
print("a" in word)
print("P" in word)
print("" in word)
print(" " in word)
print("p" in word)
```

```python
print(word)
print(id(word))
```

```python
a = 100
b = 100
print(id(a))
print(id(b))
print(a is b) ## Important is not equal ==
```

```python
a = 1000
b = 1000
print(id(a))
print(id(b))
print(id(1000))
print(a is b) ## Important 'is' not equal '=='
```

```python
a = 1000
print(a == 1000)
print(a is 1000)
```
