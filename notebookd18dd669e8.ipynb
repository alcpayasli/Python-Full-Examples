# Format Method

```python
print("{} is easy language".format('Python'))
```

```python
text = "{} and {} are fast cars"
print(text.format('Ferrari', "Lamborg"))
```

```python
print('{2} {0} {1}'.format('Python'," Java", 'C++'))
```

```python
print('{2} {0} {1} {0} {0}'.format('Python'," Java", 'C++'))
```

```python
print("{language} classes.".format(language = 'Python'))
```

```python
print("{:>15}".format('python'))
```

```python
print("|{:>15}|".format('python'))
```

```python
print("|{:<15}|".format('python'))
```

```python
print("|{:^15}|".format('python'))
```

```python
for i,j in enumerate(dir(str)):
    if (i %3 == 0):
        print("\n", end = "")
    print('|{:^20}|'.format(j), end = '')
```

```python
print('{:s}'.format("Characters"))
```

```python
print("{:c}".format(65))
```

```python
print("{:d}".format(65))
```

```python
print("{:o}".format(65))
```

```python
print("{:x}".format(65))
```

```python
print("{:X}".format(65))
```

```python
for i in range(20):
    print("{:d} {:10X} {:10b}".format(i,i,i))
```

```python
print("{:.3f}".format(50))
```

```python
print("{:,}".format(123456789))
```
