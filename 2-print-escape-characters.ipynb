## Sep Parameter

```python
print("word1", "word2", sep = ",")
```

```python
print("www", "google", "com", sep = ".")
```

```python
print("Machine Learning", "Deep Learning", "Data Science", sep = "-")
```

```python
print("one", "two", "three", "four code", sep = " code\t")
```

```python
print(1,2,3,4,5,sep = " " "***" " ")
```

```python
print("a", "b", sep = None) # None means default value
```

```python
print("First Line", "Second Line", "Third Line", sep = "\n")
```

## End Parameter

```python
print("Hello World.")
print("This language is Python.")
```

```python
print("Hello World.", end = "-")
print("This language is Python.")
```

## File Parameter

```python
file = open("try.txt", "w")
print("Hello world.", file=file) # Writing in file.
file.close()
```

```python
file = open("try.txt")
print(file.read()) # Reading file
file.close()
```
