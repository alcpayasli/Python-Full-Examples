# Accessing Elements

```python
fruits = ['apple', 'orange','banana', 'strawberry']
print(fruits[0])
print(fruits[-1])
print(fruits[:2])
print(fruits[::-1])
```

```python
for i in fruits:
    print(i)
```

```python
for i, j in enumerate(fruits, 1):
    print("{}. fruit: {}".format(i,j))
```

# Change and Add Elements of List

```python
colors = ['red', 'blue', 'green', 'black']
colors[3] = 'white'
print(colors)
```

```python
list_ = [3,4,5,6,7]
list_[:len(list_)] = [10,11,12,13,14]
print(list_)
list_[::] = [2,3]
print(list_)
```

# Combine, Delete Lists and Removing Elements from Lists

```python
compiler_languages = ["C", "C++", "C#","Java"]
interpreting_languages = ["Python", "Perl","Ruby"]
programming_languages = compiler_languages + interpreting_languages
print(programming_languages)
```

```python
del(compiler_languages)
```

```python
del(interpreting_languages[-1])
print(interpreting_languages)
```
