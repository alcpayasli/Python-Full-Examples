```python
phone_numbers= {"ahmet oz" : "0532532 32 32",
                "mehmet su": "0543 54342 42",
                "seda naz": "0533 533 33 33",
                "eda ala" : "0212212 12 12"}
name = input('Enter the name:')
if name in phone_numbers:
    print('{} is phone number: {}'.format(name, phone_numbers[name]))
else:
    print('There is no {} in phone numbers.'.format(name))
```

# Dictionary Comprehensions

```python
letters = 'abcçdefgğhıijklmnoöprsştuüvyz'
```

```python
dictionary
```

```python
names =["ahmet", "mehmet", "fırat", "zeynep", "selma", "abdullah", "cem"]
names_count = {i: len(i) for i in names}
```
