# max-min

```python
import numpy as np
number_list = list(np.random.randint(200,500,20))
```

```python
print(*number_list, sep = ',')
```

```python
max(number_list)
```

```python
names = ["ahmet", "mehmet", "necla", "sedat", 
          "abdullah", "gıyaseddin","sibel", "can","necmettin", "savaş", "özgür"]
```

```python
max(names,key=len)
```

```python
min(number_list)
```

```python
min(names, key = len)
```
