## sort

```python
members =['Ahmet', 'Mehmet', 'Ceylan', 'Seyhan', 
          'Mahmut', 'Zeynep', 'Abdullah', 'Kadir', 'Kemal', 'Kamil', 
          'Selin', 'Senem', 'Sinem', 'Tayfun', 'Tuna', 'Tolga']
members.sort()
print(members)
```

```python
numbers = [-2,0,-4,5,3,1]
numbers.sort()
print(numbers)
```

```python
letters = 'abcçdefgğhıijklmnoöprsştuüvyz'
turn = {let: letters.index(let) for let in letters}
```

```python
names =["ahmet", "ışık", "ismail", "çiğdem", "can", "şule"] 
names.sort(key = lambda x: turn.get(x[0]))
```
