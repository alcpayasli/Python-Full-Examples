```python
first_number = input("Number1:")
second_number = input("Number2:")
try:
    first_number = int(first_number)
    second_number = int(second_number)
except ValueError as error: # Orginal error message
    print(error)
else:
    try:
        print("{} / {} = {}".format(first_number, second_number, first_number/second_number))
    except ZeroDivisionError as error2:
        print(error2)
        
```
