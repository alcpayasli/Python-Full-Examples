## While Loop

```python
a = 1
while(a<10):
    print('{}. loop.'.format(a))
    a+=1
```

```python
a = 0
while(a<10):
    a+=1
    print('{}. loop.'.format(a))
```

```python
# Example Calculator

key = 1
while (key == 1):
    print("""
    Operations:
    1.Plus
    2.Minus
    3.Multiply
    4.Divide
    """)
    operations = input("Input operation number/ Press Q for exit:")
    if(operations == "q" or operations == "Q"):
        print("Exit....")
        key = 0 # Terminate loop
    elif(operations == "1"):
        num1 = int(input("Number 1:"))
        num2 = int(input("Number 2:"))
        print("Result..:", num1+num2)
    elif(operations == "2"):
        num1 = int(input("Number 1:"))
        num2 = int(input("Number 2:"))
        print("Result..:", num1-num2)
    elif(operations == "3"):
        num1 = int(input("Number 1:"))
        num2 = int(input("Number 2:"))
        print("Result..:", num1*num2)
    elif(operations == "4"):
        num1 = int(input("Number 1:"))
        num2 = int(input("Number 2:"))
        print("Result..:", num1/num2)
    else:
        print("Invalid operations.")
   
```

```python
# Example Calculator 2.way
while True:
    print("""
    Operations:
    1.Plus
    2.Minus
    3.Multiply
    4.Divide
    """)
    operations = input("Input operation number/ Press Q for exit:")
    if(operations == "q" or operations == "Q"):
        print("Exit....")
        break # Terminate loop
    elif(operations == "1"):
        num1 = int(input("Number 1:"))
        num2 = int(input("Number 2:"))
        print("Result..:", num1+num2)
    elif(operations == "2"):
        num1 = int(input("Number 1:"))
        num2 = int(input("Number 2:"))
        print("Result..:", num1-num2)
    elif(operations == "3"):
        num1 = int(input("Number 1:"))
        num2 = int(input("Number 2:"))
        print("Result..:", num1*num2)
    elif(operations == "4"):
        num1 = int(input("Number 1:"))
        num2 = int(input("Number 2:"))
        print("Result..:", num1/num2)
    else:
        print("Invalid operations.")
   
```

```python
# Consecutive Numbers with WhileLoop 
a = 1
while(a<50):
    print(a, end="-")
    a+=1
```

```python
# Even numbers with WhileLoop
a = 0
while(a<50):
    print(a, end = ",")
    a+=2
```

## For Loop

```python
letters = "abcdef"
for let in letters:
    print(let, end = ",")
```

```python
say = 0
while(say<len(letters)):
    print(letters[say], end = ",")
    say+=1
```

```python
numbers = "123456789"
for num in numbers:
    print(int(num) + 1, end = "")
```

```python
# Example Length of Password with For Loop
while True:
    password = input("Enter a password of no more than 8 characters and no less than 3 characters:")
    if not password:
        print("Password couldn't empty.")
    elif (len(password)<3 or len(password)>8):
        print("Password does not meet the criteria.")
    else:
        print("New password created.")
        break
```
