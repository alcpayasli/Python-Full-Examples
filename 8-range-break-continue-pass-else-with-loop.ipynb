# Range Function

```python
for i in range(10):
    print(i, end = ".")
```

```python
for i in range(3,20):
    print(i, end = ".")
```

```python
num = int(input("Enter the number:"))
if num in range(1,10):
    print("Number is between 1 and 10.")
else:
    print("Number is not between 1 and 10.")
```

```python
for i in range(2,20,2):
    print(i, end = ".")
    
```

```python
for i in range(1,100,3):
    print(i, end = ",")
```

```python
print(range(10))
print(*range(10))
print(*range(10), sep = "-")
```

```python
for i in range(10,0,-1):
    print(i, end = ".")
```

```python
for i in range(20,5,-3):
    print(i, end = "/")
```

# Break, Pass, Continue, Else

```python
while True:
    num = int(input("Enter the number bigger than 0."))
    if (num == 0):
        print("Exit")
        break
    elif(num<0):
        pass
    else:
        print(num)
```

```python
# Password example
password = "3456"
say =0
while True:
    user_password= input("Enter the password:")
    if not user_password:
        print("Password cannot be empty!!!")
        say+=1
    elif(user_password==password):
        print("Correct password.")
        break
    else:
        print("Incorrect password.")
        say+=1
    if(say == 3):
        print("3 consecutive incorrect or blank entries.")
        break
    
```

```python
while True:
    num = int(input("Enter the number small than 10 and positive."))
    if(num in range(0,10)):
        print("Correct number.")
        break
    elif(num<0):
        print("Enter the positive number.")
        continue
    print("Enter the number small than 10")
```
