# count

```python
word = "Python Programming Language"
word.count("P")
```

```python
word.count("P", 1)
```

```python
word.count('a', 13)
```

```python
word.count('a', 13,16)
```

```python
word.count("p")
```

```python
# Example
password = input("Enter the password different characters:")
control = True
for i in password:
    if password.count(i)>1:
        control =False
if control:
    print("Password confirmed")
else:
    print("Password cannot be same characters")
```

# rjust, ljust

```python
for i in dir(""):
    print(i.rjust(20))
```

```python
obj = "phone number"
obj.ljust(20,".")
```

```python
obj.rjust(20,".")
```
